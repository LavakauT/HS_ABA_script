```python
import os
import pandas as pd

from tfcomb import CombObj
import tfcomb.objects
```

```python
directory = '/Users/user/Desktop/all_kmer_all_gene'
dir = 'HS'
subdir = 'HSF_selected_sig'
file = 'Tak1_HS.pkl'
C = tfcomb.CombObj().from_pickle(os.path.join(directory, dir, subdir, file))
table = C.rules
table
```

```python
# get location table
locations = pd.DataFrame()
for index, row in table.iterrows():
    tf1 = row['TF1']
    tf2 = row['TF2']
    print(f'Processing: TF1={tf1}, TF2={tf2}')
    
    location = C.get_pair_locations((tf1, tf2)).as_table()
    locations = pd.concat([locations, location], axis = 0)

locations.reset_index(drop = True, inplace = True)
```

```python
print(locations)
```

```python
custom_config = {"queries": [{"distance": [10000, 10000], 
    "feature_anchor": "start", 
    "feature": "gene"}],
    "priority": True, 
    "show_attributes": "all"}
```

```python
import tfcomb.annotation
annotated = tfcomb.annotation.annotate_regions(locations, gtf = '/Users/user/Desktop/all_kmer_all_gene/MpTak_v6.1r2.gtf', config = custom_config)
```
